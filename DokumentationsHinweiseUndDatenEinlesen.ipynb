Zum Beispiel ähnlich zu folgendem:

````python

def vorwaertseinsetzen(L,b):
    """Löst für nicht singuläre untere Dreiecksmatrix Lx = b durch Vorwärtseinsetzen

    Args:
        L (np.array): Untere Dreiecksmatrix
        b (np.array): Rechte seite Vektor

    Returns:
        np.array: Löungsvektor zu Lx=b 
    """
    x = np.zeros(len(b))
    for i in np.arange(len(x)):
        x[i] =(b[i] - (x[:i]*L[i,:i]).sum())/L[i,i]
    return x

def rueckwaertseinsetzen(R,b):
    """Löst für nicht singuläre obere Dreiecksmatrix Rx = b durch Rückwärtseinsetzen

    Args:
        R (np.array): obere Dreiecksmatrix
        b (np.array): Rechte seite Vektor

    Returns:
        np.array: Löungsvektor zu Rx=b 
    """    
    x = np.zeros(len(b))
    for i in np.arange(len(x))[::-1]:
        x[i] =(b[i] - (x[i+1:]*R[i,i+1:]).sum())/R[i,i]
    return x

def LR(A):
    """Implementiert die LR zerlegung

    Args:
        A (np.array): bicht singuläre quadratische matrix

    Returns:
        : L,R untere Dreiecksmatrix und obere Dreiecksmatrix der LR zerlegung von A
    """    
    A_s = A.copy() # Matrix, die zu R umgeformt wird
    L = np.eye(*A.shape) # Matrix, die zu L aufgefüllt wird
    n = A.shape[0]
    for s in range(n-1): # Iteration durch Spalten
        A_s_old = A_s.copy()
        for i in range(s+1,n): # Iteration durch Zeilen
            l = A_s_old[i,s]/A_s_old[s,s]
            A_s[i,s+1:] = A_s_old[i,s+1:] - l * A_s_old[s,s+1:]
            L[i,s] = l
    R = np.triu(A_s)
    return L,R
````

### HDF5 Datein
`.h5` Dateiformat zum Verwalten, Verarbeiten und Speichern heterogener Daten. Es ist programmiersprachenübergreifend nutzbar.
Mit `pip install h5py` kann die entsprechende Bibliothek für Python instaliert werden.

````python
#!pip install h5py
````

````python
import h5py
````

````python
#Datein schreiben
with h5py.File('matrix_und_rechteseite.h5','w') as h5f:
    h5f['matrix'] = A
    h5f['rechteseite'] = b
````

````python
#Datein lesen
with h5py.File('matrix_und_rechteseite.h5','r') as h5f:
    A = h5f['matrix'][:]
    b = h5f['rechteseite'][:]
````
